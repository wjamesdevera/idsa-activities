## Removing Columns
* Remove Columns with more than 20% of missing/null values

```python
building_permit_nonull = building_permit.copy()
building_permit_nonull.head()
```

```python
building_permit_nonull.isnull().sum()
```

```python
print("Columns Removed")
for col in building_permit_nonull:
    sum_of_null = building_permit_nonull[col].isnull().sum()
    if (sum_of_null > (len(building_permit_nonull) * 0.2)):
        print(f"* {col}")
        del building_permit_nonull[col]
```

## Remove rows with missing values

```python
for col in building_permit_nonull:
    building_permit_nonull = building_permit_nonull.dropna(subset=[col])
    
building_permit_nonull
```

```python
building_permit_nonull.isnull().sum()
```

## Replace Current Status
* reinstated, issued, approved -> complete
* filed, planecheck, incomplete, appeal -> ongoing
* expired, withdrawn, suspend, revoked, disapproved -> cancelled

```python
building_permit_status_changed = building_permit_nonull.copy()
```

```python
building_permit_status_changed["Current Status"].unique()
```

```python
status_change = {"reinstated": "complete",
                "issued": "complete",
                "approved": "complete",
                "filed": "ongoing",
                "planecheck": "ongoing",
                "incomplete": "ongoing",
                "appeal": "ongoing",
                "expired": "cancelled",
                "withdrawn": "cancelled",
                "suspend": "cancelled",
                "revoked": "cancelled",
                "disapproved": "cancelled"}

building_permit_status_changed = building_permit_status_changed.replace({"Current Status": status_change})
building_permit_status_changed["Current Status"].unique()
```

## Remove Duplicated Permit Number
* I assumed each building permit has a unique Permit Number

```python
building_permit_no_duplicate = building_permit_status_changed.copy()
building_permit_no_duplicate.drop_duplicates(subset=["Permit Number"])
```

## Export To CSV

```python
building_permit_clean = building_permit_no_duplicate.copy()
building_permit_clean.to_csv("Building_Permit_Clean.csv")
```
